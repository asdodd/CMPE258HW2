```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import BatchNormalization
from tensorflow.keras.layers import Conv2D
from tensorflow.keras.layers import MaxPooling2D
from tensorflow.keras.layers import Activation
from tensorflow.keras.layers import Flatten
from tensorflow.keras.layers import Dropout
from tensorflow.keras.layers import Dense

from tensorflow.keras.optimizers import Adam

from tensorflow.keras.utils import to_categorical

from sklearn.datasets import make_blobs
from sklearn.model_selection import train_test_split

import numpy as np
```

```python
def build(classes=2):
  model = Sequential()

  # First Layer
  model.add(Flatten())
  model.add(Activation("relu"))
  model.add(BatchNormalization())
  model.add(Dropout(0.5)) #Regularization

  # Second Layer
  model.add(Flatten())
  model.add(Activation("relu"))
  model.add(BatchNormalization())
  model.add(Dropout(0.5)) #Regularization

  # Third Layer
  model.add(Dense(classes))
  model.add(Activation("sigmoid"))

  return model
```

```python
# Create Fake data and split into test and train (setting samples to 10000 instead of 10000000 for speed of code)
X, y = make_blobs(n_samples=10000, n_features=25,centers=2)
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=.1, train_size=.9)
```

```python
# Set epochs to 1000 and learning rate to 0.001
epochs = 1000
learningRate = 1e-3
# Changing epochs to 10 to shorten runtime
epochs = 10

# Use the Adam learning rate optimizer
opt = Adam(lr=learningRate, decay=learningRate / (epochs * 0.5))

# Build and compile the model
model = build()
model.compile(loss="binary_crossentropy", optimizer=opt,
	metrics=["accuracy"])
```

```python

print(X_train.shape)
print(y_train.shape)

numLabels = len(np.unique(y_train))
y_train1 = to_categorical(y_train, numLabels)


print(X_train.shape)
print(y_train1.shape)

```

```python


# Fit the model
H = model.fit(x = X_train, 
              y = y_train1,
              batch_size=40,
              epochs=epochs)
```
